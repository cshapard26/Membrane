```python
from ecdsa import SigningKey, SECP256k1
from eth_utils import keccak, encode_hex, to_checksum_address

# Define the private key in hexadecimal format
private_key_hex = 'c6c19ba4b9d7ea5f1f065fed3c2ff79d1c03b2d05df2b909796bceb38a209a33'

# Convert the private key to bytes
private_key_bytes = bytes.fromhex(private_key_hex)

# Create a signing key from the private key
sk = SigningKey.from_string(private_key_bytes, curve=SECP256k1)

# Get the corresponding public key
public_key = sk.verifying_key.to_string()

# Generate the Ethereum address from the public key
address = keccak(public_key).hex()[24:]  # Take the last 20 bytes (160 bits)
address = to_checksum_address('0x' + address)

# Print the results
print('Private key:', private_key_hex)
print('Public key:', encode_hex(public_key))
print('Ethereum address:', address)
```

```python
from ecdsa import SigningKey, SECP256k1
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad
import hashlib

# Generate a random private key
sk = SigningKey.generate(curve=SECP256k1)

# Get the corresponding public key
vk = sk.verifying_key.to_string()

# Encrypt a message using AES
message = b"Hello, world!"
key = hashlib.sha256(vk).digest()[:16]  # Use the first 16 bytes of the public key as the AES key
cipher = AES.new(key, AES.MODE_ECB)
encrypted_message = cipher.encrypt(pad(message, AES.block_size))

print(encrypted_message.hex())

# Now you can store `encrypted_message` somewhere safe and send `vk` along with it
# for verification or decryption
```
